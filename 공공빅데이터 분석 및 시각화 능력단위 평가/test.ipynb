```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
```

```python
# matplotlib의 글꼴 관리자에 특정 글꼴 추가
fe = fm.FontEntry(
    fname=r'font/NanumGothic.ttf',
    name='NanumGothic'
)
fm.fontManager.ttflist.insert(0, fe)

# pyplot에 matplotlib의 글꼴 관리자에 등록된 글꼴 적용
plt.rcParams.update({ 'font.family': 'NanumGothic', 'font.size': 14 })
# pyplot에서 `-`를 문자로 취급하도록 설정
plt.rcParams['axes.unicode_minus'] = False
```

<h6 style='font-weight: 700; color: #214151; background-color: #ffffff'>3. 함께 첨부된 officetel_estate.csv 파일을 DataFrame 형식의  officetel_estate 라는 변수로 불러오시오.</h6>

```python
officetel_estate = pd.read_csv('officetel_estate.csv')
officetel_estate
```

<br />

<h6 style='font-weight: 700; color: #214151; background-color: #ffffff'>4. officetel_estate DataFrame 객체의 sequence 열을 행 인덱스로 변경하여 officetel_estate 변수에 저장하시오.</h6>

```python
officetel_estate = officetel_estate.set_index('sequence')
officetel_estate
```

<br />

<h6 style='font-weight: 700; color: #214151; background-color: #ffffff'>5. officetel_estate DataFrame 객체에서 지역이 nationwide인 행만 출력하시오.</h6>

```python
officetel_estate[officetel_estate.locale.str.contains('nationwide')]
```

<br />

<h6 style='font-weight: 700; color: #214151; background-color: #ffffff'>6. officetel_estate DataFrame 객체에서 locale을 기준으로 그룹화하여 각 locale의 평당 매매가 (average_price_per_unit)의 평균을 DataFrame으로 출력하시오. (평균은 소수점 이하 3번째 자리에서 반올림하시오.)</h6>

```python
officetel_estate.groupby(officetel_estate.locale)[['lease_to_sale_ratio']].mean().round(2)
```

<br />

<h6 style='font-weight: 700; color: #214151; background-color: #ffffff'>7. officetel_estate DataFrame 객체에서 행 인덱스로 year_month, 열 인덱스로 locale, 데이터로 average_price_per_unit을 가지는 피봇 테이블을 출력하시오.</h6>

```python
officetel_estate.pivot(index='year_month', columns='locale', values='average_price_per_unit')
```

<br />

<h5 style='background-color: #114232; color: #ffffff; text-align: center; padding: 16px;'>빅데이터 분석 결과 시각화</h5>

<h6 style='font-weight: 700; color: #214151; background-color: #ffffff'>8. officetel_estate DataFrame 객체에서 nationwide의 year_month (2021년)에 대한 average_price_per_unit 변화를 나타내는 선 그래프를 pandas의 plot을 활용하여 아래와 같이 출력하시오.</h6>

<img src='./mock_image/6.png' />

```python
officetel_estate[officetel_estate.year_month.str.contains('2021')].pivot(index='year_month', columns='locale', values='average_price_per_unit').nationwide.plot()
plt.title('2021년 전국 평균 평당 매매가')
plt.xlabel('연 - 월')
plt.ylabel('평당 평균 매매가')
plt.grid(True)
```

<br />

<h6 style='font-weight: 700; color: #214151; background-color: #ffffff'>9. officetel_estate DataFrame 객체에서 seoul-northeastern, seoul-northwestern, seoul-southwestern, seoul-southeastern의 year_month (2021년)에 대한 average_price_per_unit 변화를 나타내는 막대 그래프를 pandas의 plot을 활용하여 아래와 같이 출력하시오. (figure의 사이즈는 12, 8)</h6>

<img src='./mock_image/7.png' />

```python
officetel_estate[officetel_estate.year_month.str.contains('2021')].pivot(index='year_month', columns='locale', values='average_price_per_unit')[['seoul-northeastern', 'seoul-northwestern', 'seoul-southwestern', 'seoul-southeastern']].plot.bar(rot=0, figsize=(12, 8))
plt.ylim(5000, 10000)
plt.title('2021년 서울 지역의 평균 평당 매매가')
plt.xlabel('연 월')
plt.ylabel('평균 평당 매매가')
plt.grid(True)
plt.show()
```

<br />

<h6 style='font-weight: 700; color: #214151; background-color: #ffffff'>10. officetel_estate DataFrame 객체에서 locale을 기준으로 그룹화하여 각 locale의 평당 전세가 대비 보증금 평균 (average_lease_on_deposit_price_per_unit)의 최고 값 중 서울의 막대그래프를  pandas의 plot을 활용하여 아래와 같이 출력하시오. (figure의 사이즈는 12, 8, 평균은 소수점 이하 3번째 자리에서 반올림하시오.)</h6>

<img src='./mock_image/8.png' />

```python
officetel_estate[officetel_estate.locale.str.contains('seoul')].groupby(officetel_estate.locale).average_lease_on_deposit_price_per_unit.max().round(2).plot.bar(rot=0, figsize=(12, 8))
plt.ylim(6000, 8000)
plt.title('서울 지역의 평당 전세가 대비 보증금 평균의 최고값')
plt.xlabel('지역')
plt.ylabel('평당 전세가 대비 보증금 평균')
plt.grid(True)
plt.show()
```
